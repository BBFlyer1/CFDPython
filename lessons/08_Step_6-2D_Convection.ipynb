```python
#use the 3D plotting toolkit for the following
from mpl_toolkits.mplot3d import Axes3D ## new library required for projecting 3D plots.
import numpy ##our mathematic library

from matplotlib import pyplot, cm
%matplotlib inline    
```

<p>In the following we change the number of increments in nx, ny, and nt. from the previous Step_5. We also add an v and vn variables.</p>

```python
## Declaration of variables, we've seen these before.
nx = 101
ny = 101
nt = 80
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
sigma = .2
dt = sigma * dx

x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 2, ny)

u = numpy.ones((ny, nx)) ##create a 1xn vector of 1's
v = numpy.ones((ny, nx)) ##create a 1yn vector of 1's
un = numpy.ones((ny, nx)) ##
un = numpy.ones((ny, nx)) ##

###Assign initial conditions

##set hat function I.C. : u(.5<=x<=1 && .5<=y<=1 ) is 2
u[int(0.5 / dy):int(1 / dy + 1),int(0.5 / dx):int(1 / dx + 1)] = 2 
##set hat function I.C. : u(.5<=x<=1 && .5<=y<=1 ) is 2
v[int(0.5 / dy):int(1 / dy + 1),int(0.5 / dx):int(1 / dx + 1)] = 2 
```

```python
# Plot Initial Conditions in 3D for our hat input conditions.
# The figure size parameter can be used to produce different sized images.
fig = pyplot.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
X, Y = numpy.meshgrid(x,y)

#surf = ax.plot_surface(X,Y,u[:], cmap=cm.viridis)
ax.plot_surface(X,Y,u, cmap=cm.viridis)
#now label the X and Y axis, so we can tell which is which.
ax.set_xlabel('$x$')
ax.set_ylabel('$y$');
```

```python
for n in range(nt + 1): ##loop across number of time steps
    un = u.copy()
    vn = v.copy()
#Now do it with arrays, instead of for-loops.
    u[1:, 1:] = (un[1:, 1:] -
            (un[1:, 1:] * c * dt / dx * (un[1:, 1:] - un[1:, :-1])) -
            vn[1:, 1:] * c * dt / dy * (un[1:, 1:] - un[:-1, 1:]))
    v[1:, 1:] = (vn[1:, 1:] -
            (un[1:, 1:] * c * dt / dx * (vn[1:, 1:] - vn[1:, :-1])) -
            vn[1:, 1:] * c * dt / dy * (vn[1:, 1:] - vn[:-1, 1:]))
    
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1
            
    v[0, :] = 1
    v[-1, :] = 1
    v[:, 0] = 1
    v[:, -1] = 1
```

```python
# Plot Initial Conditions in 3D for our hat input conditions.
# The figure size parameter can be used to produce different sized images.
fig = pyplot.figure(figsize=(11,7), dpi=100)
ax = fig.gca(projection='3d')
X, Y = numpy.meshgrid(x,y)

ax.plot_surface(X,Y,u, cmap=cm.viridis, rstride=2, cstride=2)
#now label the X and Y axis, so we can tell which is which.
ax.set_xlabel('$x$')
ax.set_ylabel('$y$');
```
